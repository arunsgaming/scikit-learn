Imagine we are interested in predicting penguins species based on two of their
body measurements: culmen length and culmen depth. First we want to do some
data exploration to get a feel for the data.


```python
import pandas as pd
data_penguins_species = pd.read_csv('datasets/penguins_classification.csv')
```

```python
data_penguins_species.info()
```

It has 2 numerical and 1 categorical column

```python
data_penguins_species.head()
```

```python
# different species available in the dataset
data_penguins_species['Species'].value_counts()
```

```python
# find the number of samples
print('total number of samples in dataset is ', data_penguins_species.shape[0])
```

```python
# plot histogram
_=data_penguins_species.hist(figsize=(20, 8))
```

```python
# show features distribution of each class
import seaborn as sns
columns=["Culmen Length (mm)", "Culmen Depth (mm)"]
_=sns.pairplot(
    vars=columns,
    data=data_penguins_species[:],
    hue="Species",
    plot_kws={"alpha":0.5},
    height=3,
    diag_kind="hist",
    diag_kws={"bins":10},
    )
```

```python
import matplotlib.pyplot as plt

ax=sns.scatterplot(
    x="Culmen Length (mm)",
    y="Culmen Depth (mm)",
    data=data_penguins_species,
    alpha=0.4,
    hue="Species",
)
```

Looking at the previous scatter-plot showing `"culmen length"` and `"culmen
depth"`, the species are reasonably well separated:
- low culmen length -> Adelie
- low culmen depth -> Gentoo
- high culmen depth and high culmen length -> Chinstrap

There is some small overlap between the species, so we can expect a
statistical model to perform well on this dataset but not perfectly.
